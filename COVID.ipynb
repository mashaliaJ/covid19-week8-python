```python
import pandas as pd

df = pd.read_csv("owid-covid-data.csv")  
df.head()
```

```python
df.isnull().sum()
```

```python
countries_of_interest = ['Kenya', 'Mexico', 'Norway', 'India', 'China', 'Libya' ]  
filtered_df = df[df['location'].isin(countries_of_interest)].copy() 


filtered_df['location'].value_counts()
```

```python

critical_columns = ['date', 'location', 'total_cases', 'total_deaths', 'new_cases', 'new_deaths', 'total_vaccinations',]  # Example columns


filtered_df = filtered_df.dropna(subset=critical_columns)


filtered_df.info()  
```

```python
filtered_df['date'] = pd.to_datetime(filtered_df['date'])


print(filtered_df['date'].dtype)
```

```python
filtered_df = filtered_df.fillna(0)
```

```python
filtered_df.info()  
filtered_df.head() 
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


filtered_df['date'] = pd.to_datetime(filtered_df['date'])


filtered_df = filtered_df.sort_values('date')

plt.figure(figsize=(12, 6))
sns.lineplot(
    data=filtered_df,
    x='date',
    y='total_cases',  
    hue='location',
    style='location',
    markers=True,
    dashes=False
)
plt.title('Total COVID-19 Cases Over Time by location')
plt.xlabel('Date')
plt.ylabel('Total Cases')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(
    data=filtered_df,
    x='date',
    y='total_deaths',  
    hue='location',
    style='location',
    markers=True,
    dashes=False
)
plt.title('Total COVID-19 Deaths Over Time by Location')
plt.xlabel('Date')
plt.ylabel('Total Deaths')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python

filtered_df['new_cases'] = filtered_df.groupby('location')['total_cases'].diff().fillna(0)

plt.figure(figsize=(12, 6))
sns.lineplot(
    data=filtered_df,
    x='date',
    y='new_cases',
    hue='location',
    style='location',
    markers=False,
    dashes=False
)
plt.title('Daily New COVID-19 Cases by Location')
plt.xlabel('Date')
plt.ylabel('New Cases')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
filtered_df['death_rate'] = (filtered_df['total_deaths'] / filtered_df['total_cases']) * 100  # As percentage

# Plot Death Rate Over Time
plt.figure(figsize=(12, 6))
sns.lineplot(
    data=filtered_df,
    x='date',
    y='death_rate',
    hue='location',
    style='location',
    markers=True,
    dashes=False
)
plt.title('COVID-19 Death Rate (%) Over Time')
plt.xlabel('Date')
plt.ylabel('Death Rate (%)')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Aggregate total cases by country
total_cases_by_country = filtered_df.groupby('location')['total_cases'].max().sort_values(ascending=False)

plt.figure(figsize=(10, 6))
sns.barplot(
    x=total_cases_by_country.index,
    y=total_cases_by_country.values,
    palette='viridis'
)
plt.title('Top Countries by Total COVID-19 Cases')
plt.xlabel('Country')
plt.ylabel('Total Cases')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python

heatmap_data = filtered_df.pivot_table(
    index='date',
    columns='location',
    values='total_cases'  
)

plt.figure(figsize=(12, 8))
sns.heatmap(
    heatmap_data.corr(),
    annot=True,
    cmap='coolwarm',
    linewidths=0.5
)
plt.title('Correlation Heatmap of COVID-19 Metrics by location')
plt.tight_layout()
plt.show()
```

```python

filtered_df['date'] = pd.to_datetime(filtered_df['date'])
filtered_df = filtered_df.sort_values(['location', 'date'])


filtered_df['pct_vaccinated'] = (filtered_df['people_vaccinated'] / filtered_df['population']) * 100

plt.figure(figsize=(12, 6))
sns.lineplot(
    data=filtered_df,
    x='date',
    y='people_vaccinated',  # Column for cumulative vaccinations
    hue='location',
    style='location',
    markers=False,
    linewidth=2.5
)

plt.title('Cumulative COVID-19 Vaccinations Over Time', fontsize=14)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Total Vaccinated', fontsize=12)
plt.grid(True, alpha=0.3)
plt.xticks(rotation=45)
plt.legend(title='location', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(
    data=filtered_df,
    x='date',
    y='pct_vaccinated',  # Column for % vaccinated
    hue='location',
    style='location',
    markers=False,
    linewidth=2.5
)

plt.title('Percentage of Population Vaccinated Over Time', fontsize=14)
plt.xlabel('Date', fontsize=12)
plt.ylabel('% Vaccinated', fontsize=12)
plt.grid(True, alpha=0.3)
plt.xticks(rotation=45)
plt.legend(title='location', bbox_to_anchor=(1.05, 1), loc='upper left')

# Optional: Add horizontal line for herd immunity threshold (e.g., 70%)
plt.axhline(y=70, color='red', linestyle='--', alpha=0.5, label='Herd Immunity Threshold (70%)')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
fig, ax1 = plt.subplots(figsize=(12, 6))

# Axis 1: Cumulative Vaccinations
sns.lineplot(
    data=filtered_df,
    x='date',
    y='people_vaccinated',
    hue='location',
    ax=ax1,
    legend=False,
    linewidth=2
)
ax1.set_ylabel('Total Vaccinated', color='tab:blue')
ax1.tick_params(axis='y', labelcolor='tab:blue')

# Axis 2: % Vaccinated
ax2 = ax1.twinx()
sns.lineplot(
    data=filtered_df,
    x='date',
    y='pct_vaccinated',
    hue='location',
    ax=ax2,
    style='location',
    markers=True,
    dashes=False,
    linewidth=1.5
)
ax2.set_ylabel('% Vaccinated', color='tab:orange')
ax2.tick_params(axis='y', labelcolor='tab:orange')

plt.title('Vaccinations: Absolute Count vs. % Population Coverage', fontsize=14)
plt.xlabel('Date', fontsize=12)
ax1.grid(True, alpha=0.3)

# Combine legends
lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines1 + lines2, labels1 + labels2, title='location', bbox_to_anchor=(1.2, 1))

plt.tight_layout()
plt.show()
```
